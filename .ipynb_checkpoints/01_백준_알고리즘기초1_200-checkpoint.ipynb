### 9012. 괄호

- https://www.acmicpc.net/problem/9012
- 실버4
- 220914.수.
- 시작 01:28
- 끝 01:47

```python
import sys
input_ = sys.stdin.readline

n = int(input())
for _ in range(n):
    check = False
    answer = 0
    array = input()
    for x in array:
        if answer < 0:
            print('NO')
            check = True
        if x == '(':
            answer += 1
        else:
            answer -= 1
            
    if check == True:
        continue
        
    if answer != 0:
        print('NO')
    else:
        print('YES')
```

```python
# 문제 접근 방법
# 1. 올바른 괄호 문자열인지 판별하는 함수를 만들어서 입력받은 행마다 체크
# 2. 판별 함수
#  -  ' ( ' 여는 괄호면 더해주고,
#  -  ' ) ' 닫는 괄호면 빼주고,
#  -  괄호의 수를 기록한 변수가 음수가 되거나 0이 아니면 'NO' 

import sys
input_ = sys.stdin.readline

n = int(input_())

def sol(array):
    answer = 0
    for x in array:
        if answer < 0:
            return 'NO'
        if x == '(':
            answer += 1
        else:
            answer -= 1

    if answer != 0:
        return 'NO'
    else:
        return 'YES'
            
result = []
for _ in range(n):
    array = input_().rstrip()
    result.append(sol(array))
    
for x in result:
    print(x)
```

```python
# 모범답안

from sys import stdin

n = int(input())
for _ in range(n):
    str_ = stdin.readline().strip()
    stack = 0
    for chr_ in str_:
        if chr_ == '(':
            stack += 1
        else:
            stack -= 1
            if stack < 0:
                break
    if stack == 0:
        print('YES')
    else:
        print('NO')
```

### 1874. 스택 수열

- https://www.acmicpc.net/problem/1874
- 실버2
- 220914.수.
- 시작 20:23
- 끝 22:12
- 문제를 이해 못 하겠어서 구글링해서 문제 이해만 하고 다시 품

```python
import sys
input_ = sys.stdin.readline

n = int(input())
array = [int(input()) for _ in range(n)]
array_ori = array[:]
array.sort(reverse=True)

stack = []
result = []
i = 0
stack.append(array.pop())
result.append('+')

while True:
    if not stack:
        if array:
            result = ['NO']
        break
    if stack[-1] == array_ori[i]:
        stack.pop()
        result.append('-')
        i += 1
    else:
        stack.append(array.pop())
        result.append('+')
        
print('\n'.join(result))
```

```python
import sys
input_ = sys.stdin.readline

n = int(input())
array = [int(input()) for _ in range(n)]
array_ori = array[:]
array.sort(reverse=True)

stack = []
result = []
i = 0
stack.append(array.pop())
result.append('+')

while True:
    print(i)
    if stack and stack[-1] == array_ori[i]:
        stack.pop()
        result.append('-')
        i += 1
        if not array:
            break
    elif stack and stack[-1] != array_ori[i]:
        if not array:
            result = ['NO']
            break
        stack.append(array.pop())
        result.append('+')
    elif not array:
        break

print('\n'.join(result))
```

```python
import sys
input_ = sys.stdin.readline

n = int(input())
array = [int(input()) for _ in range(n)]
array_ori = array[:]
array.sort(reverse=True)

stack = []
result = []
i = 0
stack.append(array.pop())
result.append('+')

while True:
    # 스택 값 == 입력 값인 경우
    if stack and stack[-1] == array_ori[i]:
        stack.pop()
        result.append('-')
        i += 1
    # 스택 값 != 입력 값인 경우
    elif stack and stack[-1] != array_ori[i]:
        if not array: # 매칭되는 값이 없는 경우
            result = ['NO']
            break
        stack.append(array.pop())
        result.append('+')
    # 스택에 추가할 경우
    elif not stack and array:
        stack.append(array.pop())
        result.append('+')
    # 다 쓴 경우
    elif not stack and not array:
        break

print('\n'.join(result))
```

```python
# 모범답안

import sys
def sol():
    n,*data = map(int, sys.stdin.read().split())
    now=1
    stack=list()
    ans=[]
    for tar in data:
        if tar>=now:
            stack.extend(range(now,tar))
            ans.extend('+'*(tar-now+1)+'-')
            now=tar+1
        elif stack[-1]==tar:
            ans.append('-')
            stack.pop()
        else:
            return 'NO'
    return '\n'.join(ans)
print(sol())
```

```python
array = list(input())
n = int(input())
cursor = len(array)

for i in range(n):
    print(cursor)
    command = input().split()
    
    if command[0] == 'L' and cursor > 0:
        cursor -= 1
    elif command[0] == 'D' and cursor < len(array):
        cursor += 1
    elif command[0] == 'B' and cursor > 0:
        array = array[:cursor-1] + array[cursor:]
        if cursor > 0:
            cursor -= 1
    elif command[0] == 'P':
        array = array[:cursor] + [command[1]] + array[cursor:]
        if cursor < len(array):
            cursor += 1
    print(''.join(array))
        
print(''.join(array))
print(cursor)
```

```python
# 모범 답안.. 와우

import sys

st1 = list(sys.stdin.readline().rstrip())
st2 = []

for _ in range(int(sys.stdin.readline())):
    command = list(sys.stdin.readline().split())
    if command[0] == 'L':
        if st1:
            st2.append(st1.pop())
            
    elif command[0] == 'D':
        if st2:
            st1.append(st2.pop())

    elif command[0] == 'B':
        if st1:
            st1.pop()

    else:
        st1.append(command[1])
        
st1.extend(reversed(st2))
print(''.join(st1))
```

```python
## 재도전
## 백준 제출 시에는 input_().rstrip() 해야 함

import sys
input_ = sys.stdin.readline

array = list(input())
n = int(input())

box = []

for _ in range(n):
    command = input().split()
    
    if command[0] == 'L' and array:
        box.append(array.pop())
    elif command[0] == 'D' and box:
        array.append(box.pop())
    elif command[0] == 'B' and array:
        array.pop()
    elif command[0] == 'P':
        array.append(command[1])

for x in box[::-1]:
    array.append(x)

print(''.join(array))
```

### 10845. 큐

- https://www.acmicpc.net/problem/10845
- 실버4
- 220922.목.
- 시작 19:57
- 끝 20:20

```python
n = int(input())
array = []

for i in range(n):
    command = input().rstrip().split()
    
    if command[0] == 'push':
        array.append(command[1])
    elif command[0] == 'pop':
        if not array:
            print(-1)
            continue
        print(array[0])
        array = array[1:]
    elif command[0] == 'size':
        print(len(array))
    elif command[0] == 'empty':
        if array:
            print(0)
        else:
            print(1)
    elif command[0] == 'front':
        if not array:
            print(-1)
            continue
        print(array[0])
    elif command[0] == 'back':
        if not array:
            print(-1)
            continue
        print(array[-1])
```

```python
## 깔끔한 방식?

from sys import stdin

read = stdin.readline

N = int(read())
queue = []
answer = []
command = {
    "push": queue.append,
    "pop": lambda q: q.pop(0) if q else '-1',
    "size": lambda q: str(len(q)),
    "empty": lambda q: '0' if q else '1',
    "front": lambda q: q[0] if q else '-1',
    "back": lambda q: q[-1] if q else '-1'
}

for _ in range(N):
    o = read().rstrip()
    if len(o) > 5:
        o = o.split()
        command[o[0]](o[1])
    else:
        answer.append(command[o](queue))

print('\n'.join(answer))
```

```python
n, k = map(int, input().split())
array = [i+1 for i in range(n)]
result = []

for i in range(n):
    idx = (i+1)*k-1
    
    if idx >= n:
        idx = idx-n
    
    while True:
        if array[idx] in result:
            idx += 1
        else:
            break
            
            
    result.append(array[idx])
    print(result)
    
result
```

```python
n, k = map(int, input().split())
array = [i+1 for i in range(n)]
poplist = []
result = []

for i in range(n):
    idx = (i+1)*k-1
    
    if idx >= n:
        idx = idx-n-len(poplist)
        for pl in poplist:
            array.pop(pl)
        poplist = []
            
    result.append(array[idx])
    poplist.append(idx)
    
    print(result)
    
result
```

### 10866. 덱

- https://www.acmicpc.net/problem/10866
- 실버4
- 220929.목.
- 시작 18:37
- 끝 18:53
- 쉬움

```python
n = int(input())
array = []
answer = []

for _ in range(n):
    command = input().split()
    if command[0] == 'push_front':
        array = [int(command[1])] + array
    elif command[0] == 'push_back':
        array.append(int(command[1]))
    elif command[0] == 'pop_front':
        if array:
            answer.append(array[0])
            array = array[1:]
        else:
            answer.append(-1)
    elif command[0] == 'pop_back':
        if array:
            answer.append(array.pop())
        else:
            answer.append(-1)
    elif command[0] == 'size':
        answer.append(len(array))
    elif command[0] == 'empty':
        answer.append(0 if array else 1)
    elif command[0] == 'front':
        if array:
            answer.append(array[0])
        else:
            answer.append(-1)
    elif command[0] == 'back':
        if array:
            answer.append(array[-1])
        else:
            answer.append(-1)

for x in answer:
    print(x)
```
